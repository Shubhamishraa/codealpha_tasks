```python
import gradio as gr
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# FAQ Dataset
faq = {
    "What is Python?": "Python is a popular programming language.",
    "What is AI?": "AI stands for Artificial Intelligence.",
    "Who created Python?": "Python was created by Guido van Rossum.",
    "What is Machine Learning?": "Machine Learning is a branch of Artificial Intelligence.",
    "What is Data Science?": "Data Science is the process of extracting useful insights from data.",
    "What is Deep Learning?": "Deep Learning is a subset of Machine Learning.",
    "What is NLP?": "NLP stands for Natural Language Processing.",
    "What is Cloud Computing?": "Cloud Computing provides computing services over the internet.",
    "What is Web Development?": "Web Development is the process of creating websites and web applications.",
    "What is Data Analytics?": "Data Analytics is the process of analyzing data for better decisions.",
    "Do you offer internships?": "Yes, we offer internships in AI, Python, Data Science and Web Development.",
    "How can I apply for an internship?": "You can apply through our careers page.",
    "Do you provide certificates?": "Yes, certificates are provided after successful completion of the course.",
    "Do you provide placement assistance?": "Yes, we provide placement assistance after course completion.",
    "Do you provide online training?": "Yes, we provide online training with live sessions.",
    "What services do you offer?": "We offer AI, Web Development, Python Training and Data Analytics services.",
    "Do you provide remote services?": "Yes, we provide services remotely worldwide.",
    "How can I contact support?": "You can contact support via email or phone.",
    "What are your business hours?": "Our business hours are Monday to Friday, 9:00 AM to 6:00 PM.",
    "How can I enroll in a course?": "You can enroll by visiting our website or contacting our support team."
}

# Prepare Data
questions = list(faq.keys())
answers = list(faq.values())

# TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words="english")
faq_vectors = vectorizer.fit_transform(questions)

# Chatbot Function
def chatbot(user_question):
    if not user_question.strip():
        return "Please enter a question."

    user_vector = vectorizer.transform([user_question])

    similarity = cosine_similarity(user_vector, faq_vectors)

    best_match = similarity.argmax()
    score = similarity[0][best_match]

    if score >= 0.30:
        return answers[best_match]
    else:
        return "Sorry, I couldn't find an answer. Please ask another question related to our FAQs."

# Gradio Interface
interface = gr.Interface(
    fn=chatbot,
    inputs=gr.Textbox(
        lines=2,
        placeholder="Ask your question here..."
    ),
    outputs=gr.Textbox(label="Answer"),
    title="🤖 FAQ Chatbot using NLP",
    description="This chatbot uses TF-IDF and Cosine Similarity to answer Frequently Asked Questions.",
    theme="soft"
)

interface.launch(share=True)
```
